#### Com certeza você pelo menos já ouviu falar na técnica de identificar **microexpressões** faciais para, a partir delas, conseguir obter informações de sentimentos de uma pessoa, tais como *alegria*, *tristeza*, etc.

#### Hoje, com a inovação batendo à porta a cada instante, a tecnologia nos presenteia com vários modelos que podem extrair **pontos de referência faciais** e **microexpressões**. Vamos falar de uma dessas inovações: *PY-FEAT*

#### E o que seria o PY-FEAT? O nome significa **"Python Facial Expression Analysis Toolbox"**. É literalmente uma biblioteca cheia de ferramentas para análise de microexpressões faciais.

#### Agora vamos começar! Antes de tudo, o que precisamos é instalar essa biblioteca. Vamos lá.

```python
#Instalando a biblioteca PY-FEAT

pip install py-feat
```

#### Com a biblioteca instalada, podemos agora criar o detector. Faremos da seguinte forma:

```python
#Criando o detector de Microexpressões

from feat import Detector

detector = Detector()
```

#### Finalmente, começaremos nossa análise. Para isso, basta uma linha de código de python e o caminho da imagem a ser analisada:

```python
#Setando a imagem a ser analisada

result = detector.detect_image('/content/img1.jpg')
```

#### Com isso, o sistema já armazena as informações da imagem. Para visualizar essas informações, utilizaremos um método chamado "*plot_detections*"

```python
#Plotando os resultados da análise

result.plot_detections(add_titles=False);
```

```python
result.plot_detections(faces='aus', muscles=True, add_titles=False);
```

#### E o resultado é incrível! Além de detectar a emoção contida na imagem, "raiva" (anger), o *PY-FEAT* ainda nos traz uma visualização de **quais conjuntos de músculos faciais estão sendo ativados para compor aquela expressão**.

#### Para tornar um pouco mai interessante, podemos deixar o usuário final enviar o caminho da imagem, tornando ainda mais desafiador para o sistema. Vamos ver como será o comportamento!

#### Para teste, a imagem será essa:

img2.jpg

```python
#Criando um Input e analisando a imagem enviada
#Para recriar o sistema, basta executar essa célula de código e colocar um caminho para uma imagem ao seu desejo.

imagem = input('Envie uma imagem para eu analisar as expressões faciais: ')

result = detector.detect_image(imagem)

result.plot_detections(add_titles=False);

result.plot_detections(faces='aus', muscles=True, add_titles=False);
```

#### E mais uma vez o resultado é surpreendente! Ele identificou, ainda, mais de uma emoção, trazendo mais precisão à análise.

#### Concluímos que o *PY-FEAT* é uma ferramenta no mínimo incrível de análise de microexpressões faciais, trazendo informações precisas, com riqueza de detalhes, visualização e muito mais. Vitor Metaforando deve se preocupar?!
